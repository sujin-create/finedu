# **맞춤형 카테고리에 해당하는 동전주의 관련 기사를 보여주고 종목을 고르도록 올바른 투자습관을 제공**

- naver 뉴스창를 활용해 추천 동전주들의 뉴스기사를 크롤링 해온다 

```python
from ast import keyword
import requests
from bs4 import BeautifulSoup
```

```python
#pip install pyautogui
```

```python
print("투자전 우리는 해당 주식이 오를것인가?를 고민해야합니다. 묻지마투자는 이제 그만하세요!!")

# 종목별로 관련 뉴스를 먼저 보여줌으로써, 묻지마 투자 습관을 버리고 공부후투자로 습관을 바로잡음
for interest in interests['종목명'].values:
  resp = requests.get(f"https://search.naver.com/search.naver?where=news&sm=tab_jum&query=동전주{interest}") # interest coin input으로 받음
  print('<<<{0}>>>뉴스를 읽고 투자를 결정하세요'.format(interest))
  html = resp.text
  soup = BeautifulSoup(html,"html.parser")
  links = soup.select('.news_tit')
  for link in links: #각 link는 <a>내용 </a> 의 html
      title = link.text # 태그 안에 텍스트요소 가져온다! -> 제목
      url = link.attrs['href'] # href=""의 속성값을 가져온다. -> 링크
      print(title, url) # 제목, 링크 크롤링 완료!
  print("\n")
```

# **사용자는 구매한 종목에 대한 뉴스 기사를 보며 매도 타이밍을 공부한다**

- 사용자가 구매한 종목에 대한 뉴스기사를 크롤링을 해서 가져온다
- 매도 여부를 결정하기 위해 장 시작시간 10분전인 8시50분에 관련 뉴스를 보여주어 종목의 흐름을 파악한다


```python
User_coin = ['인디에프', 'SK증권'] # 사용자가 매수한 동전주에 인디에프, sk증권이 있다고하자
```

```python
for stock in User_coin:
  print(stock)
  resp = requests.get(f"https://search.naver.com/search.naver?where=news&sm=tab_jum&query=동전주{stock}") # 사용자가 가진 동전주의 뉴스를 가져옴
  print('<<<{0}>>>뉴스를 읽고 매도를 결정하세요'.format(stock))
  html = resp.text
  soup = BeautifulSoup(html,"html.parser")
  links = soup.select('.news_tit')
  for link in links: #각 link는 <a>내용 </a> 의 html
      title = link.text # 태그 안에 텍스트요소 가져온다! -> 제목
      url = link.attrs['href'] # href=""의 속성값을 가져온다. -> 링크
      print(title, url) # 제목, 링크 크롤링 완료!
  print("\n")
```
